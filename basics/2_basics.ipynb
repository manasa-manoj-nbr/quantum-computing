## Hadamard Gate

```python
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler

qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()


sampler = Sampler()
job = sampler.run(qc)
result = job.result()
print(result.quasi_dists[0])
qc.draw('mpl')
```

## CNOT

```python
qc1 = QuantumCircuit(2)
qc1.h(0)
qc1.cx(0,1)
qc1.measure_all()
qc1.draw('mpl')
```

## Paulli Gates

```python
qc2 = QuantumCircuit(1)
qc2.x(0)
qc2.measure_all()

job2 = sampler.run(qc2)
result2 = job2.result()
print(result2.quasi_dists[0])
qc2.draw('mpl')

```
